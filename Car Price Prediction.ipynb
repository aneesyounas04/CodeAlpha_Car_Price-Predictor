```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('car data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.duplicated().sum()
```

```python
ds_cleaned=df
```

```python
ds_cleaned.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
car_name=LabelEncoder()
df['Car_Name']=car_name.fit_transform(df['Car_Name'])
df
```

```python
df['Fuel_Type'].unique()
```

```python
Fuel_Type=LabelEncoder()
df['Fuel_Type']=Fuel_Type.fit_transform(df['Fuel_Type'])
df
```

```python
Selling_type=LabelEncoder()
df['Selling_type']=Selling_type.fit_transform(df['Selling_type'])
df
```

```python
Transmission=LabelEncoder()
df['Transmission']=Transmission.fit_transform(df['Transmission'])
df
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
input_data=df.iloc[:,:-1]
output_data=df['Selling_Price']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
input_data=pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
input_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
model=LinearRegression()
model.fit(X_train,y_train)
model.score(X_train,y_train)*100,model.score(X_test,y_test)*100
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Use R square to compare the test and pred values
r2=r2_score(y_test,y_pred)
r2
```

```python
# Now use adjusted R square
n=X_test.shape[0]
p=X_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1))/(n-p-1)
print("Adjusted R Square",adjusted_r2)
```

```python
mean_squared_error(y_test,model.predict(X_test)),mean_absolute_error(y_test,model.predict(X_test))
```

```python
model1=Lasso(alpha=0.05)
model1.fit(X_train,y_train)
model1.score(X_train,y_train)*100,model1.score(X_test,y_test)*100
```

```python
mean_squared_error(y_test,model1.predict(X_test)),mean_absolute_error(y_test,model1.predict(X_test))
```

```python
model2=Ridge(alpha=5)
model2.fit(X_train,y_train)
model2.score(X_train,y_train)*100,model2.score(X_test,y_test)*100
```

```python
mean_squared_error(y_test,model2.predict(X_test)),mean_absolute_error(y_test,model2.predict(X_test))
```
